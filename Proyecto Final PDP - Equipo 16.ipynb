#1_ Identificación del problema

##Identificación del problema
###Problemática
La sobrepoblación, o superpoblación, se define como: “Exceso de individuos de una especie o de un conjunto de especies en un espacio determinado.” (RAE, Diccionario de la lengua española, https://dle.rae.es/superpoblaci%C3%B3n#NL64F0D)

###Consecuencias
La sobrepoblación provoca hambruna (escasez generalizada de alimentos), desigualdad social, agotamiento de los recursos naturales, degradación del medio ambiente, desempleo, encarecimiento del coste de la vida, contaminación de los recursos naturales (hídricos, medio ambiente, degradación de la calidad del aire), degradación de la biodiversidad, enfermedades y pandemias, y a la larga, extinción de la especie.

###Objetivo
Entender la problemática global de sobrepoblación, por medio del análisis del crecimiento demográfico mundial a lo largo del tiempo, para buscar soluciones y/o recomendaciones de mitigación para la problemática.

###Justificación
Debido a la relación natural entre la población humana y el medio ambiente, y específicamente por la influencia que tiene la primera sobre el segundo, es muy importante encontrar medidas que controlen el crecimiento de la población, o bien, que disminuya o desacelere su crecimiento, para poder conseguir que la población crezca de una manera sostenible, evitando el deterioro de su entorno natural y la degradación de sus recursos naturales.

###Método
Para entender la problemática de la sobrepoblación, primero es necesario comprender el comportamiento del crecimiento de la población a nivel mundial, a lo largo del tiempo. Este proyecto se centra en ese análisis, en el período de tiempo entre 1960 y 2018, contemplando diferentes aspectos, como lo son el tamaño total de la población por país y la tasa de crecimiento en un período de tiempo específico, bajo diferentes criterios, como población total, población por género (masculino/femenino), mayores de 65 años, menores de 14 años y densidad de población por unidades de área (km²). Este análisis se hará por medio del planteamiento de preguntas específicas. 

##Investigación del problema
Wikipedia (página web, https://es.wikipedia.org/wiki/Superpoblaci%C3%B3n_humana) nos dice que la superpoblación o sobrepoblación es un fenómeno que se produce cuando una elevada densidad de población, puede provocar un empeoramiento del entorno, una disminución en la calidad de vida o situaciones de hambre y conflictos. También puede aplicarse a cualquier otra especie que alcance niveles críticos en su número de individuos.

La sobrepoblación proviene de la superación, por una especie animal, de los límites de sostenibilidad del biotopo (en ecología, es un área de condiciones ambientales uniformes que provee espacio vital a un conjunto de flora y fauna) que habita. Esto es, una especie demanda más alimento, produce más residuos y exige más espacio del que el biotopo puede darle o aceptar sin sacrificar el futuro de otras especies que habitan en él. La sobrepoblación es tanto una cuestión de espacio, como de limitación de recursos, y sobre todo, como consecuencia directa, de extinción de las especies que cohabitan la región superpoblada con la especie humana.

De acuerdo con “Sostenibilidad para Todos”, la humanidad ya alcanzo los 8.000 millones de personas viviendo en el planeta. Una cifra nunca antes vista que no para de aumentar de manera exponencial, y esta sobrepoblación se presenta como un tremendo reto para el mundo en la gestión de los recursos, en el afronte de las desigualdades y los desequilibrios para conseguir un desarrollo sostenible.

Las principales causas de la superpoblación son el descenso de la tasa de mortalidad gracias a la medicina principalmente, progresos en la producción de alimentos y migraciones y concentración urbana.

```python
# Carga el archivo World_Population.csv
df_total = pd.read_csv(ruta + "World_Population_limpio.csv")
df_total
```

```python
# Se calcula la población mundial por año y se establece el índice por año
df_world_pop = df_total.groupby('Year')['Total_population'].sum().reset_index()
df_world_pop = df_world_pop.rename(columns = {'Total_population': 'world_population'})
df_world_pop = df_world_pop.set_index('Year', drop = True)
```

```python
# Se preparan los datos para calcular la tasa de crecimiento con respecto al año previo

# 1. Se copian los datos de población mundial por año sin índice
df_world_pop_previous_year = df_world_pop.copy().reset_index()
# 2. Se desplaza el año en una unidad para
df_world_pop_previous_year['Year'] = df_world_pop_previous_year['Year'] + 1
# 3. Colocarlo como el valor de la población mundial del año previo
df_world_pop_previous_year = df_world_pop_previous_year.rename(columns = {'world_population' : 'world_population_previous_year'})
# 4. Se indexan los datos por año
df_world_pop_previous_year = df_world_pop_previous_year.set_index('Year', drop = True)
# 5. Se mezclan los datos de población mundial con el de la población del año previo
df_world_pop = pd.merge(df_world_pop, df_world_pop_previous_year, on=['Year'], how='left')
# 6. Como se desplazó un año se generó un valor NaN que hay que limpiar 
#df_world_pop['world_population_previous_year'] = df_world_pop['world_population_previous_year'].fillna(0)
# 7. Por el valor NaN el tipo de dato habría dejado de ser int64
#diccionario_de_conversion = { 'world_population_previous_year': 'int64' }
# 8. Se realiza el cast correspondiente
#df_world_pop = df_world_pop.astype(diccionario_de_conversion)
```

```python
# Se calcula la tasa de crecimiento por año, con respecto a la población del año previo
df_world_pop['pct_rate_increase'] = ((df_world_pop['world_population'] - df_world_pop['world_population_previous_year'])/df_world_pop['world_population_previous_year']) * 100
```

```python
# Se combina el total de la poblacion mundial por año con el Dataset original
df_temp = pd.merge(df_total, df_world_pop, on = ['Year'], how='left' )

# Se establece el índice por País y Año para preparar el cálculo de porcentaje de población mundial
df_temp = df_temp.set_index(["Country", "Year"], drop=True)

# Se calcula el porcentaje de población mundial de cada País, Año
df_temp['pct_world_population'] = ((df_temp["Total_population"] / df_temp["world_population"]) * 100)
df_temp
```

```python
# Funciones y variables, auxiliares

years = df_clean['Year'].unique();
countries_list = df_clean['Country'].unique();
def format_number(item):
    return '{:,.2f}'.format(item)

def format_Per_cent(item):
    return str(item) + ' %'

def population_by_year(dataframe, years, order = False):
    columns  = ['Year', 'Country', 'Total_population']
    years_list, countries, population = [], [], []
    for year in years:
        item = dataframe.loc[year].sort_values('Total_population', ascending=order).iloc[0]
        years_list.append(year)
        countries.append(item.name)
        population.append(item[0])
    return pd.DataFrame(list(zip(years, countries, population)), columns= columns).set_index('Year')

def create_intervals_raise(population_list, years):
    columns = ['Interval', 'Raise_population']
    interval, raise_list = [], []
    for i , n in enumerate(population_list):
        if i != len(population_list) - 1 :
            interval.append(str(years[i]) + ' - ' + str(years[i + 1]))
            raise_list.append(population_list[i + 1] - population_list[i])
    return pd.DataFrame(list(zip(interval, raise_list)), columns= columns).set_index('Interval')

def population_raise_mean_country(list_country, order = False):
    data = []
    for country in countries_list:
        population_list = pd.DataFrame(list_country[country].sort_index()).reset_index()['Total_population']
        data.append({'Country': country, 'Population_mean_raise': format_number(create_intervals_raise(population_list, years)['Raise_population'].mean())})
    return pd.DataFrame(data).set_index('Country').sort_values('Population_mean_raise', ascending= order).head(1)

def inverso_porcentaje(item):
    return 100 - item
```
